```python
# I have imported all needed libraries for work

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#I have open my file to start working on it .

df= pd.read_csv('earthquake.csv')

#I checked  how many rows got my document and basic information as : mean,std, max, min  etc.
df.describe()

# I didn't want to  check whole data I took just the top 5
df.head()

# I have created  histogram of each numerical column in my DB which gave me result of 12 histograms out of 22 columns
df.hist(bins=10, figsize=(10,10)) 
plt.show()
```

```python
# created size 3 plot figure getting ready for my bar graph
plt.figure(figsize=(3,3)) 

# I use status column and created bar graph with reviewed, automatic statuses 
df['status'].value_counts().plot(kind='bar' , color=['r', 'cyan'], fontsize=9)
#labeled my vertical x for client to know which one was  done automatically by pc , and which one did analyst added/edited
plt.xlabel('Status of Earthquakes', color='r', fontsize=10)
#I've called my horizontal y number of Earthquakes that you can have rough number how many been analyzed by pc vs data analyst
plt.ylabel('Number of Earthquakes', color='g', fontsize=10)
#created title of my graph
plt.title('Earthquakes Information', color='b', fontsize=20)
plt.show()

#Automatic - Earthquake Center receives seismic data from hundreds of stations in real-time.
# we are able to use a computer program to generate initial locations

#Reviewed - Analyst reviewed locations.The automated system misses some earthquakes and many arrivals,
#  so analysts scan for these and add them manually.

#The graph shows the distribution of earthquake data by status, with the "reviewed" status represented by the red bar and the "automatic" status represented by the cyan bar.
#  From the graph, it appears that the majority of earthquakes in the dataset were  reviewed and potentially edited by a data analyst , with a smaller number  automatically processed by a computer.
# The number of earthquakes in the "automatic" category is relatively low compared to the number in the "reviewed" category. 
# The title of the graph is "Earthquakes Information", and the x-axis is labeled "Status of Earthquakes" while the y-axis is labeled "Number of Earthquakes" . 
# That gives us a rough idea of the number of earthquakes that have been analyzed by the computer versus the number that have been analyzed and edited by a data analyst.
```

```python
#created x and y values to hold latituted and longtiude as var
x=np.array([df.latitude])
y=np.array([df.longitude])



# started plotting my graph , setted a marker as o and green color of it
plt.plot(x,y, color='g', marker='o')
#gave the name to my x axis latitude and set color,fontsize and fontweight
plt.xlabel('Latitude N ', color='mediumblue', fontsize=15,fontweight='bold') 
#gave the name to my y axis  and added extras for font
plt.ylabel('Longitude W ',color='orange', fontsize=15,fontweight='bold') 
#I have set graph title 
plt.title('Locations of Earthquakes by Latitude and Longitude', fontsize=20, fontweight='bold')
#grid on axis y  and x to make it look more detailed
plt.grid(axis='y', color='red', linewidth=1)
plt.grid(axis='x', color='red', linewidth=1)

#The graph shows the locations of earthquakes as represented by their latitude and longitude coordinates.
#The graph shows the locations of earthquakes by plotting the latitude and longitude points on the x and y axis respectively.
#The gridlines are added to make the graph more detailed and easy to read, the grid is in red color with a linewidth of 1 on both x and y axis.
#Overall, this graph provides a visual representation of the distribution of earthquake locations worldwide based on their latitude and longitude coordinates.
```

```python
# I have assing var count_c to hold my groupby via place. I grouped them/count them
count_c=df.groupby(['place']).count() 
# I took TOP 5 Highest risk places of earthquakes
top_5 =count_c.nlargest(5,'time')


# I used top_5 to hold places and amount of time the earthquake happened 
top_5[['time']].plot(kind='bar',y='time', color=['darkred','firebrick','indianred','lightcoral','rosybrown'])
# I've labeled the x axis as Places where earthquakes happened and changed color and size to be easily visible
plt.xlabel('Places' ,fontsize=15 , color='magenta')
# I've labeled the y axis as  number of times the earthquakes did happened , I added color for txt and size
plt.ylabel('Number of Eartquakes' , color='darkblue', fontsize=15)
# Put title for my graph  and set color, size , fontweight
plt.title('TOP 5 Earthquakes Places ', fontsize=20 , color='darkgreen' , fontweight='bold')
# I have set visible to false for legend to make it more clear for the user
plt.legend().set_visible(False)
plt.show()


#The graph represents the top 5 places where earthquakes occurred the most, as determined by the number of times an earthquake was recorded in each place.
#The places are represented on the x-axis and the number of times an earthquake occurred in each place is represented on the y-axis.
#Overall, this graph provides a visual representation of the top 5 places where earthquakes occurred the most, based on the data in the dataset.
```

```python
# I used top_5 to hold places and amount of time the earthquake happened 
top_5[['time']].plot(kind='pie',y='time', autopct='%1.1f%%')
# Put title for my graph  and set color, size , fontweight
plt.title('TOP 5 Earthquakes Places', fontsize=20 , color='darkred' , fontweight='bold')
#I have 'hidden' my label because it was in way of the graph
plt.ylabel("number of earthquakes").set_visible(False)

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5) )
plt.show()

#The graph is a pie chart that represents the top 5 places where earthquakes occurred the most, as determined by the number of times an earthquake was recorded in each place.
#The legend of the graph is moved to the center left and is anchored to the graph so it doesn't interfere with the visualization.
#Overall, this graph provides a visual representation of the proportion of earthquakes that occurred in the top 5 places, and it's easy to compare the difference between them.
```
