```python
# Machine Learning cluster

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering
```

```python
#load data set


data = pd.read_csv('Wholesale customers data.csv')
data.head()
```

```python
# Normalizing data set , use normalize for unsupervied machine learning

data_norm=normalize(data)

#created the standard scaler and sign to dataFrame
data_norm=pd.DataFrame(data_norm, columns=data.columns)

data_norm.head()
```

```python
#Dendogram

plt.figure(figsize=(10,10))
plt.title('Dendograms')
dend = shc.dendrogram(shc.linkage(data_norm,method='ward'))
```

```python
plt.figure(figsize=(10,10))
plt.title('Dendograms')
dend = shc.dendrogram(shc.linkage(data_norm,method='ward'))

plt.axhline(y=6, color='r', linestyle='--')
```

```python
# train model

cluster= AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
cluster.fit_predict(data_norm)
```

```python
## scatter plot  to show clusters 

plt.figure(figsize=(5,5))

plt.scatter(data_norm['Milk'], data_norm['Grocery'], c=cluster.labels_ , cmap='rainbow')
```

```python
plt.figure(figsize=(5,5))

plt.scatter(data_norm['Fresh'], data_norm['Frozen'], c=cluster.labels_ , cmap='rainbow')
```
